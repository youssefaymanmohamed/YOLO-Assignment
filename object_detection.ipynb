### Directory Structure

- YOLOv5â€™s directory structure includes several important folders:

    - data/: Contains datasets and configuration files. For example, data/images/ is where your images should be placed for detection.

    - models/: Includes pre-trained model configurations and weights. You can find YOLOv5 model definitions here, like yolov5s.yaml.

    - runs/: Stores the output of your detection runs. After running detection, the results are saved in directories such as runs/detect/exp/.

    - utils/: Contains utility scripts used by YOLOv5, including functions for parsing arguments, logging, and drawing bounding boxes.

    - data.yaml: A configuration file that defines dataset paths and class names for training and evaluation.

    - detect.py: The main script used for performing object detection. You can run this script with various arguments to detect objects in images or videos.

```python
import cv2
import numpy as np
```

```python
# Load YOLOv5 model (choose a model size: small 's', medium 'm', large 'l', or extra-large 'x')
model_5s = torch.hub.load('ultralytics/yolov5', 'yolov5s')  # or 'yolov5m', 'yolov5l', 'yolov5x'

# Path to the video
video_path = Path('vid1.mp4')

# Open the video file
cap = cv2.VideoCapture(str(video_path))

# Check if the video was opened successfully
if not cap.isOpened():
    print("Error: Could not open video.")
    exit()

# Define the allowed classes (e.g., 'person', 'car')
allowed_classes = ['person', 'car']

# Process the video frame by frame
while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break  # Exit the loop if no frames are left

    # Convert the frame to RGB (YOLOv5 uses RGB)
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Perform inference on the frame
    results = model_5s(rgb_frame)

    # Get all detections (unfiltered)
    unfiltered_results = results.pandas().xyxy[0]   # 0 is the index of the first frame

    # Create a copy for unfiltered results
    unfiltered_frame = frame.copy() # Copy the frame to draw unfiltered results

    # Draw bounding boxes for unfiltered results
    for _, row in unfiltered_results.iterrows():
        xmin, ymin, xmax, ymax = int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax']) # Bounding box coordinates
        label = f"{row['name']} {row['confidence']:.2f}" # Label for the bounding box
        color = (255, 0, 0)  # Blue color for unfiltered results
        cv2.rectangle(unfiltered_frame, (xmin, ymin), (xmax, ymax), color, 2) # Draw the bounding box
        cv2.putText(unfiltered_frame, label, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2) # Draw the label

    # Filter results to display only specific classes
    filtered_results = unfiltered_results[unfiltered_results['name'].isin(allowed_classes)] # Filter results by allowed classes

    # Draw bounding boxes for filtered classes
    for _, row in filtered_results.iterrows():
        xmin, ymin, xmax, ymax = int(row['xmin']), int(row['ymin']), int(row['xmax']), int(row['ymax']) # Bounding box coordinates
        label = f"{row['name']} {row['confidence']:.2f}" # Label for the bounding box 
        color = (0, 255, 0)  # Green color for filtered results
        cv2.rectangle(frame, (xmin, ymin), (xmax, ymax), color, 2) # Draw the bounding box 
        cv2.putText(frame, label, (xmin, ymin - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2) # Draw the label

    # Stack both frames side-by-side for comparison
    combined_frame = cv2.vconcat([unfiltered_frame, frame]) # Combine both frames vertically

    # Resize the combined frame to make it smaller (e.g., 50% of original size)
    resized_frame = cv2.resize(combined_frame, None, fx=0.3, fy=0.3, interpolation=cv2.INTER_LINEAR) 

    # Display the resized frame
    cv2.imshow("Unfiltered (up) vs Filtered (down)", resized_frame) # Display the frame

    # Break the loop if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Release the video capture and close windows
cap.release()
cv2.destroyAllWindows()
```
